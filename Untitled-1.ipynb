UNEMPLOYMENT PREDICTION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from datetime import datetime
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import *
from sklearn.linear_model import *
from sklearn.metrics import *
from sklearn.ensemble import *

dataset = pd.read_csv("Unemployment in India.csv")
dataset.head()
```

```python
sns.countplot(data=dataset,x='Area')
```

```python
lb=LabelEncoder()
dataset['Region']=lb.fit_transform(dataset['Region'])
dataset['Frequency']=lb.fit_transform(dataset['Frequency'])
dataset['Area']=lb.fit_transform(dataset['Area'])
dataset.dropna(subset=['Date'],inplace=True)
```

```python
date=pd.to_datetime(dataset['Date'],errors='coerce')
dataset['Date']=date.apply(lambda x: x.timestamp())
```

```python
dataset['Region'].fillna(dataset['Region'].mean(),inplace=True)
dataset['Frequency'].fillna(dataset['Frequency'].mean(),inplace=True)
dataset['EstimatedUnemploymentRate'].fillna(dataset['EstimatedUnemploymentRate'].mean(),inplace=True)
dataset['EE'].fillna(dataset['EE'].mean(),inplace=True)
dataset['EstimatedLabourParticipation'].fillna(dataset['EstimatedLabourParticipation'].mean(),inplace=True)
dataset['Area'].fillna(dataset['Area'].mean(),inplace=True)

print(dataset.isnull().sum())
```

```python
dataset.index.is_unique
```

```python
sns.pairplot(dataset)
```

```python
print(dataset['EstimatedUnemploymentRate'].skew())
sns.distplot(dataset['EstimatedUnemploymentRate'])
```

```python
dataset['EstimatedUnemploymentRate']=np.log(dataset['EstimatedUnemploymentRate'])
print(dataset['EstimatedUnemploymentRate'].skew())
```

```python
X = dataset.drop(columns='EstimatedUnemploymentRate',axis=1)
Y = dataset[['EstimatedUnemploymentRate']]

print(X)
print(Y)

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3,random_state=0)
print(xtrain.shape)
print(xtest.shape)

ytrain = ytrain.replace((np.inf, -np.inf, np.nan)).reset_index(drop=True)
ytest = ytest.replace((np.inf, -np.inf, np.nan)).reset_index(drop=True)

model = LinearRegression()
model.fit(xtrain,ytrain)
print("Model :",model)

prediction = model.predict(xtest)
print("Predicted Values :",prediction)
```

```python
print("Linear Regression predictions")
score = r2_score(ytest,prediction)*100
mae = mean_absolute_error(ytest,prediction)
mse = mean_squared_error(ytest,prediction)

print(f"R2 score      : {score:.2f}")
print(f"MAE           : {mae:.2f}")
print(f"MSE           : {mse:.2f}")  
```

RANDOM FOREST REGRESSOR

```python
model1 = RandomForestRegressor(n_estimators=10000,n_jobs=1000)
model1.fit(xtrain,ytrain)
print("Model 1 :",model1)

p1 = model1.predict(xtest)
print(p1)
```

```python
print("RandomForestRegressor predictions")
score1 = r2_score(ytest,p1)*100
mae1 = mean_absolute_error(ytest,p1)
mse1 = mean_squared_error(ytest,p1)

print(f"R2 score      : {score1:.2f}")
print(f"MAE           : {mae1:.2f}")
print(f"MSE           : {mse1:.2f}")  
```
